```python
#Tic Tac Toe with difficulty selector (Easy/Hard)
import random
import ipywidgets as widgets
from IPython.display import display, clear_output

#Global variaables
board = [[" "]*3 for _ in range(3)]
buttons=[[None]*3 for _ in range(3)]
turn = "X"
difficulty = "Hard"

#Check winner
def check_winner(mark):
  return any(all(board[r][c]== mark for c in range(3)) for r in range(3)) or\
  any(all(board[r][c] == mark for r in range(3)) for c in range(3)) or\
  all(board[i][i]== mark for i in range(3)) or\
  all(board[i][2-i]== mark for i in range(3))

#Check if board is full
def is_full():
  return all(board[r][c] != " " for r in range(3) for c in range(3))

#MinMAx algorithms
def minimax(is_maximizing):
  if check_winner("O"): return 1
  if check_winner("X"): return -1
  if is_full(): return 0

  if is_maximizing: #Ai's Turn
      best_score = -999
      for r in range(3):
        for c in range(3):
          if board[r][c]==" ":
            board[r][c] = "O"
            score = minimax(False)
            board[r][c] = " "
            best_score = max(best_score,score)
      return best_score
  else: #Player's Turn
      best_score = 999
      for r in range(3):
        for c in range(3):
          if board[r][c]==" ":
            board[r][c]="X"
            score = minimax(True)
            board[r][c]= " "
            best_score = min(best_score,score)
      return best_score

#Ai moves depends on difficulty
def ai_move():
  global difficulty
  if difficulty == "Easy":
    #Random move
    empty = [(r, c) for r in range(3) for c in range(3) if board[r][c] == " "]
    if empty:
      r, c = random.choice(empty) # Use random.choice for single selection
    else:
      return # No empty cells, game is over

  else: # Difficulty is Hard (Minimax)
    best_score = -999
    move = None
    for r in range(3):
      for c in range(3):
        if board[r][c]==" ":
          board[r][c]= "O"
          score = minimax(False)
          board[r][c] = " "
          if score > best_score:
            best_score = score
            move = (r, c)
    if move:
      r, c = move
    else:
      return # No possible moves (board is full)

  #Play the chosen move
  board[r][c] = "O"
  buttons[r][c].description = "O"
  buttons[r][c].disabled = True

  if check_winner("O"):
    print("AI wins!")
    disable_all()
    return
  if is_full():
    print("It's a tie!")
    disable_all()
    return

#handle player move
def handle_click (r, c , btn):
  global turn
  if board[r][c] == " " and turn == "X":
    board[r][c] = "X"
    btn.description = "X"
    btn.disabled = True
    if check_winner("X"):
      print("player win!")
      disable_all()
      return
    if not is_full():
      ai_move()
    else:
        print("It's a tie!")
        disable_all()
        return


#Disable all buttons
def disable_all():
  for r in range(3):
    for c in range(3):
      buttons[r][c].disabled = True

#Clear the board display
def clear_board():
    for r in range(3):
        for c in range(3):
            buttons[r][c].description = " "
            buttons[r][c].disabled = False

#Reset game
def reset_game(btn):
  global board, buttons, turn
  board = [[" "]*3 for _ in range(3)]
  turn = "X"
  clear_output()
  clear_board()
  print(f"New game started! You are 'X'.Difficulty:{difficulty}")

#Difficulty slector
def set_difficulty(change):
  global difficulty
  difficulty = change["new"]
  print(f"Difficulty set to: {difficulty}")

#create board with buttons
def create_board():
  for r in range(3):
    for c in range(3):
      btn = widgets.Button(description = " ",layout = widgets.Layout(width = "60px", height = "60px"))
      btn.on_click(lambda btn, r=r, c=c: handle_click(r, c, btn))
      buttons[r][c] = btn

  # Display the board after creating all buttons
  board_rows = [widgets.HBox(buttons[r]) for r in range(3)]
  display(widgets.VBox(board_rows))

  reset_btn = widgets.Button(description = "Reset Game", button_style="warning")
  reset_btn.on_click(reset_game)
  display(reset_btn)

  diff_selector = widgets.Dropdown(
      options = ["Easy","Hard"], # Changed 'option' to 'options'
      value = difficulty,
      description = "Difficulty:"
  )
  diff_selector.observe(set_difficulty,names = "value")
  display(diff_selector)

#START THE GAME
create_board()
print(f"Game started! you are 'X'.Difficulty:{difficulty}")
```
