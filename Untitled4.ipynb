```python
#Initializing and importing library
!pip install SpeechRecognition pyttsx3 pyaudio
import speech_recognition as sr
import pyttsx3
import datetime
import webbrowser
import os
import sys
import platform
```

```python
class VoiceAssistant:
    def __init__(self):
        # Initialize text-to-speech engine
        self.engine = pyttsx3.init()
        self.engine.setProperty('rate', 175)  # Speed of speech
        self.engine.setProperty('volume', 0.9)  # Volume (0.0 to 1.0)

        # Initialize speech recognizer
        self.recognizer = sr.Recognizer()
        self.recognizer.energy_threshold = 4000
        self.recognizer.dynamic_energy_threshold = True

    def speak(self, text):
        """Convert text to speech"""
        print(f"Assistant: {text}")
        self.engine.say(text)
        self.engine.runAndWait()

    def listen(self):
        """Listen to microphone and convert speech to text"""
        with sr.Microphone() as source:
            print("\nListening...")
            self.recognizer.adjust_for_ambient_noise(source, duration=0.5)
            try:
                audio = self.recognizer.listen(source, timeout=5, phrase_time_limit=10)
                print("Processing...")

                # Use Google Speech Recognition
                command = self.recognizer.recognize_google(audio).lower()
                print(f"You said: {command}")
                return command

            except sr.WaitTimeoutError:
                print("No speech detected")
                return None
            except sr.UnknownValueError:
                self.speak("Sorry, I couldn't understand that. Please try again.")
                return None
            except sr.RequestError:
                self.speak("Sorry, there was an error with the speech service.")
                return None


```

```python
def get_time(self):
        """Get current time"""
        now = datetime.datetime.now()
        time_str = now.strftime("%I:%M %p")
        self.speak(f"The current time is {time_str}")

def get_date(self):
        """Get current date"""
        now = datetime.datetime.now()
        date_str = now.strftime("%B %d, %Y")
        day = now.strftime("%A")
        self.speak(f"Today is {day}, {date_str}")

def search_web(self, query):
        """Search the web using default browser"""
        search_query = query.replace("search for", "").replace("search", "").strip()
        if search_query:
            url = f"https://www.google.com/search?q={search_query}"
            webbrowser.open(url)
            self.speak(f"Here are the search results for {search_query}")
        else:
            self.speak("What would you like me to search for?")

def open_website(self, query):
        """Open specific websites"""
        if "youtube" in query:
            webbrowser.open("https://www.youtube.com")
            self.speak("Opening YouTube")
        elif "google" in query:
            webbrowser.open("https://www.google.com")
            self.speak("Opening Google")
        elif "github" in query:
            webbrowser.open("https://www.github.com")
            self.speak("Opening GitHub")
        else:
            self.speak("Sorry, I don't have that website in my list")

def open_application(self, query):
        """Open applications based on OS"""
        system = platform.system()

        try:
            if "notepad" in query and system == "Windows":
                os.system("notepad.exe")
                self.speak("Opening Notepad")
            elif "calculator" in query:
                if system == "Windows":
                    os.system("calc.exe")
                elif system == "Darwin":  # macOS
                    os.system("open -a Calculator")
                elif system == "Linux":
                    os.system("gnome-calculator &")
                self.speak("Opening Calculator")
            elif "browser" in query or "chrome" in query:
                if system == "Windows":
                    os.system("start chrome")
                elif system == "Darwin":
                    os.system("open -a 'Google Chrome'")
                elif system == "Linux":
                    os.system("google-chrome &")
                self.speak("Opening browser")
            else:
                self.speak("Sorry, I cannot open that application")
        except Exception as e:
            self.speak(f"Sorry, I couldn't open that application. {str(e)}")

def process_command(self, command):
        """Process the voice command and execute appropriate action"""
        if command is None:
            return True

        # Greetings
        if any(word in command for word in ["hello", "hi", "hey"]):
            self.speak("Hello!NIshant, How can I help you today?")

        # Time
        elif "time" in command:
            self.get_time()

        # Date
        elif "date" in command or "day" in command:
            self.get_date()

        # Search web
        elif "search" in command or "google" in command:
            self.search_web(command)

        # Open websites
        elif "open" in command and any(site in command for site in ["youtube", "google", "github"]):
            self.open_website(command)

        # Open applications
        elif "open" in command:
            self.open_application(command)

        # Help
        elif "hi" in command or "what can you do" in command:
            help_text = """I can help you with the following:
            - Tell the time: say 'what time is it'
            - Tell the date: say 'what is the date'
            - Search the web: say 'search for python tutorials'
            - Open websites: say 'open YouTube' or 'open Google'
            - Open applications: say 'open calculator' or 'open notepad'
            - Exit: say 'exit', 'quit', or 'goodbye'
            """
            self.speak(help_text)

        # Exit
        elif any(word in command for word in ["exit", "quit", "goodbye", "bye", "stop"]):
            self.speak("Goodbye! Have a great day!")
            return False

        else:
            self.speak("I'm not sure how to help with that. Say 'help' to see what I can do.")

        return True

def run(self):
        """Main loop to run the voice assistant"""
        self.speak("Voice Assistant activated. How can I help you?")

        while True:
            command = self.listen()
            should_continue = self.process_command(command)

            if not should_continue:
                break

def main():
    """Main function to start the assistant"""
    print("=" * 50)
    print("Voice Controlled Assistant")
    print("=" * 50)
    print("\nMake sure your microphone is connected and working.")
    print("Say 'help' to see available commands.")
    print("Say 'exit' to quit the assistant.\n")

    try:
        assistant = VoiceAssistant()
        assistant.run()
    except KeyboardInterrupt:
        print("\n\nAssistant stopped by user.")
    except Exception as e:
        print(f"\nAn error occurred: {e}")
        print("Please make sure you have installed: pip install SpeechRecognition pyttsx3 pyaudio")

if __name__ == "__main__":
    main()
```
