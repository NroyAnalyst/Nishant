```python
import json
import pandas as pd
from pandas import json_normalize
from pathlib import Path

# Path to your JSON file
json_path = Path("/content/daikibo-telemetry-data.json")

# Load JSON
with json_path.open("r") as f:
    data = json.load(f)

# Flatten JSON
# (works whether the top level is a list of records or a dict)
df_flat = json_normalize(data)

# Save to CSV
csv_path = Path("daikibo-telemetry-data_flat.csv")
df_flat.to_csv(csv_path, index=False)

print("Flattened data preview:")
print(df_flat.head())
print("CSV written to:", csv_path)
```
