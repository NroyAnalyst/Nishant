```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
#step 1: Load dataset
data=pd.read_csv("/content/HousingData.csv")
```

```python
#Step 2: Hndle missing value
data=data.dropna()
```

```python
#step 3: Feature(X) and target(y)
X= data.drop("MEDV",axis=1)
Y=data["MEDV"]
```

```python
#step 4: Normalize feature
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#step 5: Train-test-split
X_train, X_test, Y_train, Y_test=train_test_split(X_scaled, Y, train_size=0.8, test_size=0.2, random_state =42)
```

```python
#step 6: Train Linear Regression model on normalized data
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
#step 7: predict on test data
Y_pred = model.predict (X_test)
```

```python
#step 8: Evaluate Model
mae = mean_absolute_error(Y_test, Y_pred)
mse = mean_squared_error(Y_test, Y_pred)
rmse =np.sqrt(mse)
r2=r2_score(Y_test, Y_pred)
```

```python
print(" Model Evaluation Metrics (with Normalized Data):")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared Score (R²): {r2:.2f}")
```

```python
plt.figure(figsize=(8,6))
plt.scatter(Y_test, Y_pred, color="blue", edgecolors="k", alpha=0.7)
plt.plot([min(Y_test), max(Y_test)], [min(Y_test), max(Y_test)],
 color="red", linestyle="--", linewidth=2)
plt.xlabel("Actual MEDV ($1000s)")
plt.ylabel("Predicted MEDV ($1000s)")
plt.title("Actual vs Predicted House Prices (Normalized Features)")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
